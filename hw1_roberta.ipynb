```python
import torch
import numpy as np
from transformers import AutoTokenizer, AutoModel
from datasets import load_dataset
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
from umap import UMAP
from collections import defaultdict
import gc
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Device: {device}")
if torch.cuda.is_available():
    print(f"Memory: {torch.cuda.get_device_properties(0).total_memory / 1e9:.2f} GB")
```

```python
model_name = "roberta-base"

tokenizer = AutoTokenizer.from_pretrained(model_name, use_fast=True)
model = AutoModel.from_pretrained(model_name, output_hidden_states=True)
model.to(device)
model.eval()

print(f"Model type: {model.config.model_type}")
print(f"Hidden size: {model.config.hidden_size}")
print(f"Num hidden layers: {model.config.num_hidden_layers}")
```

```python
dataset = load_dataset("yelp_polarity", split="train")
dataset = dataset.shuffle(seed=42).select(range(4000))
texts = dataset["text"]; labels = np.array(dataset["label"])
labels = np.array(dataset['label'])
print(f"Dataset size: {len(texts)}")
print(f"Classes: {np.unique(labels)}")
```

```python
def extract_layer_representations(texts, model, tokenizer, batch_size=8, max_length=256):
    all_layer_representations = defaultdict(list)

    model.eval()
    with torch.no_grad():
        for i in range(0, len(texts), batch_size):
            batch_texts = texts[i:i+batch_size]

            inputs = tokenizer(
                batch_texts,
                padding=True,
                truncation=True,
                max_length=max_length,
                return_tensors="pt"
            ).to(device)

            outputs = model(**inputs)
            hidden_states = outputs.hidden_states

            for layer_idx, layer_output in enumerate(hidden_states):
                cls_embeddings = layer_output[:, 0, :].cpu().numpy()
                all_layer_representations[layer_idx].append(cls_embeddings)

            del inputs, outputs, hidden_states
            torch.cuda.empty_cache()

            if (i // batch_size + 1) % 10 == 0:
                print(f"Processed {i+batch_size}/{len(texts)} samples")

    for layer_idx in all_layer_representations:
        all_layer_representations[layer_idx] = np.vstack(all_layer_representations[layer_idx])

    return all_layer_representations
layer_representations = extract_layer_representations(texts, model, tokenizer)
print(f"Extracted representations for {len(layer_representations)} layers")
print(f"Shape of layer 0: {layer_representations[0].shape}")
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    np.arange(len(labels)),
    labels,
    test_size=0.2,
    random_state=42,
    stratify=labels
)
results = {
    'layer': [],
    'train_acc': [],
    'test_acc': [],
    'f1_score': [],
    'mean_norm': [],
    'std_norm': []
}
```

```python
for layer_idx in sorted(layer_representations.keys()):
    X = layer_representations[layer_idx]

    X_train_layer = X[X_train]
    X_test_layer = X[X_test]
    y_train_labels = np.array(labels)[X_train]
    y_test_labels = np.array(labels)[X_test]

    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train_layer)
    X_test_scaled = scaler.transform(X_test_layer)

    clf = LogisticRegression(max_iter=1000, random_state=42, C=1.0)
    clf.fit(X_train_scaled, y_train_labels)

    train_pred = clf.predict(X_train_scaled)
    test_pred = clf.predict(X_test_scaled)

    train_acc = accuracy_score(y_train_labels, train_pred)
    test_acc = accuracy_score(y_test_labels, test_pred)
    f1 = f1_score(y_test_labels, test_pred, average='binary')

    norms = np.linalg.norm(X, axis=1)
    mean_norm = np.mean(norms)
    std_norm = np.std(norms)

    results['layer'].append(layer_idx)
    results['train_acc'].append(train_acc)
    results['test_acc'].append(test_acc)
    results['f1_score'].append(f1)
    results['mean_norm'].append(mean_norm)
    results['std_norm'].append(std_norm)

    print(f"Layer {layer_idx}: Test Acc={test_acc:.4f}, F1={f1:.4f}")

    gc.collect()
```

```python
%matplotlib inline
```

```python
# ПЕРЕД fig, axes = plt.subplots(2, 2, figsize=(15, 12))
print("DEBUG INFO:")
print(f"results dict keys: {results.keys()}")
print(f"Number of layers: {len(results['layer'])}")
print(f"Test accuracy values: {results['test_acc']}")
print(f"Layer values: {results['layer']}")
```

```python
import pandas as pd
results_df = pd.DataFrame(results)
print("\n" + "="*60)
print("RESULTS TABLE")
print("="*60)
print(results_df.to_string(index=False))
print("="*60)
top3_indices = np.argsort(results_df['test_acc'].values)[-3:][::-1]
top3_layers = results_df.iloc[top3_indices]['layer'].values
print(f"\nTop 3 layers: {top3_layers}")

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for idx, layer in enumerate(top3_layers):
    print(f"Processing UMAP for layer {layer}...")

    X_layer = layer_representations[int(layer)]

    sample_size = min(800, len(X_layer))
    sample_indices = np.random.choice(len(X_layer), size=sample_size, replace=False)
    X_sample = X_layer[sample_indices]
    y_sample = labels[sample_indices]

    print(f"  Sample size: {len(X_sample)}, Classes: {np.unique(y_sample)}")

    reducer = UMAP(n_components=2, random_state=42, n_neighbors=15, min_dist=0.1)
    X_reduced = reducer.fit_transform(X_sample)

    print(f"  UMAP shape: {X_reduced.shape}")

    scatter = axes[idx].scatter(
        X_reduced[:, 0],
        X_reduced[:, 1],
        c=y_sample,
        cmap='coolwarm',
        alpha=0.6,
        s=30,
        edgecolors='k',
        linewidth=0.5
    )

    test_acc = results_df[results_df['layer'] == layer]['test_acc'].values[0]
    axes[idx].set_title(f'Layer {int(layer)}\nTest Acc: {test_acc:.4f}',
                        fontsize=12, fontweight='bold')
    axes[idx].set_xlabel('UMAP 1', fontsize=10)
    axes[idx].set_ylabel('UMAP 2', fontsize=10)
    axes[idx].grid(True, alpha=0.3)

    cbar = plt.colorbar(scatter, ax=axes[idx])
    cbar.set_label('Class', fontsize=10)

plt.tight_layout()
plt.savefig('umap_visualization.png', dpi=300, bbox_inches='tight')
plt.show()
print("✓ UMAP visualization saved!")
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

axes[0, 0].plot(results['layer'], results['test_acc'], marker='o', linewidth=2, markersize=8, color='#2E86AB')
axes[0, 0].axhline(y=np.mean(results['test_acc']), color='red', linestyle='--', alpha=0.5, label='Mean')
axes[0, 0].set_xlabel('Layer Index', fontsize=12)
axes[0, 0].set_ylabel('Test Accuracy', fontsize=12)
axes[0, 0].set_title('Test Accuracy by Layer', fontsize=14, fontweight='bold')
axes[0, 0].grid(True, alpha=0.3)
axes[0, 0].legend()

axes[0, 1].plot(results['layer'], results['f1_score'], marker='s', linewidth=2, markersize=8, color='#A23B72')
axes[0, 1].axhline(y=np.mean(results['f1_score']), color='red', linestyle='--', alpha=0.5, label='Mean')
axes[0, 1].set_xlabel('Layer Index', fontsize=12)
axes[0, 1].set_ylabel('F1 Score', fontsize=12)
axes[0, 1].set_title('F1 Score by Layer', fontsize=14, fontweight='bold')
axes[0, 1].grid(True, alpha=0.3)
axes[0, 1].legend()

axes[1, 0].plot(results['layer'], results['mean_norm'], marker='^', linewidth=2, markersize=8, color='#F18F01')
axes[1, 0].set_xlabel('Layer Index', fontsize=12)
axes[1, 0].set_ylabel('Mean L2 Norm', fontsize=12)
axes[1, 0].set_title('Representation Magnitude by Layer', fontsize=14, fontweight='bold')
axes[1, 0].grid(True, alpha=0.3)

gap = np.array(results['train_acc']) - np.array(results['test_acc'])
axes[1, 1].plot(results['layer'], gap, marker='D', linewidth=2, markersize=8, color='#C73E1D')
axes[1, 1].axhline(y=0, color='black', linestyle='-', alpha=0.3)
axes[1, 1].set_xlabel('Layer Index', fontsize=12)
axes[1, 1].set_ylabel('Train-Test Gap', fontsize=12)
axes[1, 1].set_title('Generalization Gap by Layer', fontsize=14, fontweight='bold')
axes[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('layer_analysis.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
layer_groups = {
    'Initial (0-2)': [0, 1, 2],
    'Middle (3-4)': [3, 4],
    'Final (5-6)': [5, 6]
}

group_stats = []
for group_name, layers in layer_groups.items():
    group_df = results_df[results_df['layer'].isin(layers)]
    stats = {
        'Group': group_name,
        'Mean Test Acc': group_df['test_acc'].mean(),
        'Std Test Acc': group_df['test_acc'].std(),
        'Mean F1': group_df['f1_score'].mean(),
        'Best Layer': group_df.loc[group_df['test_acc'].idxmax(), 'layer']
    }
    group_stats.append(stats)
```

```python
group_stats_df = pd.DataFrame(group_stats)
print("\n" + "="*60)
print("LAYER GROUP COMPARISON")
print("="*60)
print(group_stats_df.to_string(index=False))
print("="*60)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

colors = ['#6A4C93', '#1982C4', '#8AC926']
positions = np.arange(len(group_stats_df))
bars = ax.bar(positions, group_stats_df['Mean Test Acc'],
               yerr=group_stats_df['Std Test Acc'],
               color=colors, alpha=0.7, capsize=10, edgecolor='black', linewidth=1.5)

ax.set_xlabel('Layer Group', fontsize=12, fontweight='bold')
ax.set_ylabel('Mean Test Accuracy', fontsize=12, fontweight='bold')
ax.set_title('Performance Comparison Across Layer Groups', fontsize=14, fontweight='bold')
ax.set_xticks(positions)
ax.set_xticklabels(group_stats_df['Group'], fontsize=11)
ax.grid(True, alpha=0.3, axis='y')

for i, (bar, val) in enumerate(zip(bars, group_stats_df['Mean Test Acc'])):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height + 0.01,
            f'{val:.4f}', ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.savefig('group_comparison.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
print("\n" + "="*80)
print("CONCLUSIONS")
print("="*80)

best_layer = results_df.loc[results_df['test_acc'].idxmax()]
worst_layer = results_df.loc[results_df['test_acc'].idxmin()]

print(f"\n1. BEST PERFORMING LAYER:")
print(f"   Layer {int(best_layer['layer'])} achieved the highest test accuracy of {best_layer['test_acc']:.4f}")
print(f"   F1 Score: {best_layer['f1_score']:.4f}")

print(f"\n2. LAYER GROUP ANALYSIS:")
for _, row in group_stats_df.iterrows():
    print(f"   {row['Group']}: Mean Acc = {row['Mean Test Acc']:.4f} (±{row['Std Test Acc']:.4f})")

print(f"\n3. KEY FINDINGS:")
middle_acc = group_stats_df[group_stats_df['Group'] == 'Middle (3-4)']['Mean Test Acc'].values[0]
initial_acc = group_stats_df[group_stats_df['Group'] == 'Initial (0-2)']['Mean Test Acc'].values[0]
final_acc = group_stats_df[group_stats_df['Group'] == 'Final (5-6)']['Mean Test Acc'].values[0]

if middle_acc > initial_acc and middle_acc > final_acc:
    print("   - Middle layers show the best performance, confirming HiProbe-VAD hypothesis")
    print("   - Initial layers capture low-level features insufficient for classification")
    print("   - Final layers may be overly specialized for language modeling task")
elif final_acc > initial_acc and final_acc > middle_acc:
    print("   - Final layers show the best performance")
    print("   - Deep representations are most discriminative for this task")
else:
    print("   - Initial layers show competitive performance")
    print("   - Lower layers already capture sufficient semantic information")

print(f"\n4. PERFORMANCE RANGE:")
print(f"   Accuracy range: {worst_layer['test_acc']:.4f} - {best_layer['test_acc']:.4f}")
print(f"   Relative improvement: {((best_layer['test_acc'] - worst_layer['test_acc']) / worst_layer['test_acc'] * 100):.2f}%")

print(f"\n5. RECOMMENDATION:")
print(f"   Use layer {int(best_layer['layer'])} for optimal feature extraction on this task")
print(f"   Top-3 layers: {top3_layers} can be used for ensemble approaches")

print("="*80)
```

##Выводы

Гипотеза HiProbe-VAD частично подтверждается: пик информативности находится не на самом последнем слое, а в финальных. Лучший результат даёт 9-й слой (test acc ≈ 0.92, F1 ≈ 0.92); близки к нему 8 и 11. При этом 12-й слой немного уступает 9-му — т.е. максимум не совпадает с самым последним.
Если смотреть усреднённые метрики по группам слоёв, они монотонно растут с глубиной: Initial (0–2) ≪ Middle (3–4) < Final (5–6). Это подтверждает тренд «чем глубже, тем полезнее», с заметным скачком после 3–4 слоёв.
